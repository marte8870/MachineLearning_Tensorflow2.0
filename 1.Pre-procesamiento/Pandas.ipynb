```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris


iris = load_iris()
iris['feature_names'] 
```

- Asignar nombres a columnas

```python
names = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm',
       'Species']
```

- Cargar datos y etiquetas

```python
df = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= names)
```
